```python

# Example 1: Simple Line Plot

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a plot
plt.plot(x, y)

# Add labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
# Example 2: Bar Plot

import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 1, 10]

# Create a bar plot
plt.bar(categories, values)

# Add labels
plt.xlabel('Categories')
plt.ylabel('Values')

# Display the plot
plt.show()
```
